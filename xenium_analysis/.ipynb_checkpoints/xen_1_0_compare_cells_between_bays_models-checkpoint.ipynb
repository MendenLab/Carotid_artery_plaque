# Defining functions

```python
### =================================================================

### FILTER low confidence transcripts & create anndata from Baysor output
def generate_anndata(panel_sample_name,bays_model_name,seg,seg_stats,nucleus_polygon_pixel_metrics,confidence_thr=0.8,
                     assignment_confidence_thr=0.5,qv_thr=20,print_filtering_results=True,plot_confidence_distr=False):

    import logging
    logging.basicConfig(format='%(asctime)s - %(message)s', datefmt='%d-%b-%y %H:%M:%S')
    logging.getLogger().setLevel(logging.INFO)

    ### Columns names of 'segmentation.csv' (as 'seg' df variable) dataframe => 
    #   Based on the description if the output columns found at https://github.com/kharchenkolab/Baysor
    '''
    confidence: probability of a molecule to be real (i.e. not noise)
    cell: id of the assigned cell. Value "" corresponds to noise.
    cluster: id of the molecule cluster
    assignment_confidence: confidence that the molecule is assigned to a correct cell
    is_noise: shows whether molecule was assigned to noise (it equals true if and only if cell == "")
    ncv_color: RGB code of the neighborhood composition coloring
    '''
    
    #### FILTER TRANSCRCIPTS WITH LOW CONFIDENCE
    ## Filter result of baysor segmentation based on the threshold values 
    seg_true=seg[(~seg['cell'].isna())&\
                (seg['confidence']>confidence_thr)&\
                (seg['assignment_confidence']>assignment_confidence_thr)&\
                (seg['qv']>qv_thr)] 


    #### EXTRACT COUNT MATRIX OF TRANSCRIPTS
    ## Extract the transcript counts for each gene for each cell 
    a=seg_true.groupby('cell').apply(lambda x: (x['gene'].value_counts(dropna=False)))

    l=[]
    for cell_series in a.index.get_level_values('cell').unique():
        # Rename 'count' column to gene => This will be the index column with the gene names in the final concatenated dataframe
        c=a[cell_series]
        c.rename(cell_series, inplace=True)
        l.append(c)

    ## Concatenate the list of pd.series of gene counts for each cell
    count_df=pd.concat(l,axis=1,join='outer').T.fillna(0)
    count_df=count_df.rename(index={'gene': 'cell_id'})


    #### CREATE ADATA.OBS & ADATA.VAR DATAFRAME
    ## Set cell_id as index for seg_stats df to make it uniform with other df-s
    seg_stats=seg_stats.rename(columns={'cell': 'cell_id'})
    seg_stats=seg_stats.set_index('cell_id')

    ## Extract list of True cell names => intersection of filtered cells (count_df) and 
    #  cells with non-zero polygon mask (nucleus_polygon_pixel_metrics contains only cells with non-zero masks see function 
    #  "create_cell_bound_df()" which created cell_bound_df)
    true_cells_ind=list(set(count_df.index)&set(nucleus_polygon_pixel_metrics.index))

    ## Add metrics of cell mask DAPI intensities and subset only to true cells 
    mask_stat_cols=['mean','median','max','mode']
    seg_stats_=seg_stats.copy()
    seg_stats_.loc[true_cells_ind,mask_stat_cols]=nucleus_polygon_pixel_metrics.loc[true_cells_ind,mask_stat_cols].values
    seg_stats_=seg_stats_.rename(columns={'mean':'mean_DAPI_intensity','median':'median_DAPI_intensity','max':'max_DAPI_intensity','mode':'mode_DAPI_intensity'})
    
    # Calculate log10 values of intensity metrics as they have a large range
    for metric in mask_stat_cols:
        colname='_'.join([metric,'DAPI_intensity'])
        log_colname='_'.join(['log',metric,'DAPI_intensity'])
        seg_stats_[log_colname]=np.log10(seg_stats_[colname])

    # Add sample and patient metadata
    seg_stats_['original_sample']=panel_sample_name
    seg_stats_['panel']=panel_sample_name.split('_')[0]
    seg_stats_['patient']=panel_sample_name.split('_')[1]
    seg_stats_['condition']=panel_sample_name.split('_')[-1]

    
    ## Calculate the area of the cell masks 
    #  => segmentation_stats.csv 'area' column: area of the convex hull around the cell molecules, not the cell masks' area
    cell_mask_areas=nucleus_polygon_pixel_metrics['raw_vals'].apply(lambda x: x.shape[0])
    seg_stats_['cell_mask_area']=cell_mask_areas[true_cells_ind]

   
    ## Create adata.obs dataframe
    obs=seg_stats_.loc[true_cells_ind,:].copy()

    # Create adata.var dataframe
    var=pd.DataFrame(index=count_df.columns)


    #### CREATE ANNDATA OBJECT
    # Extract the count matrix of true cells from count_df
    from scipy import sparse
    X=count_df.loc[true_cells_ind,:].values
    adata=ad.AnnData(X, obs=obs, var=var, dtype='float64')
    adata.X=sparse.csr_matrix(adata.X)
    adata.layers['raw_counts']=adata.X.copy()
    adata.layers['norm_by_area']=sparse.csr_matrix(adata.layers['raw_counts'].multiply(1/(adata.obs['cell_mask_area'].values.reshape(-1,1))))


    #### ADD ADDITIONAL METADATA & CALCULATE BASIC STATISTICS
    ## Add noise information
    pct_noise=sum(seg['cell'].isna())/len(seg)
    adata.uns['pct_noise']=pct_noise

    ## Add raw DAPI intensity values of cell masks
    #adata.uns['raw_DAPI_intensities']=nucleus_polygon_pixel_metrics['raw_vals']

    # Calculate some basic statistics   
    adata.obs['n_counts']=adata.X.sum(axis=1).A1
    adata.obs['n_genes']=(adata.X>0).sum(axis=1).A1
    adata.var['n_counts']=adata.X.sum(axis=0).A1
    adata.var['n_cells']=(adata.X>0).sum(axis=0).A1

    ## Print basic information about dataset
    if print_filtering_results==True:  
        logging.info(f"Dataset: {panel_sample_name} - {bays_model_name}")
        logging.info(f'Number of raw transcripts: {len(seg)}')
        logging.info(f"Number of transcripts deemed as noise: {sum(seg['cell'].isna())} - {round(pct_noise*100,2)}%")
        logging.info(f'Number of transcripts kept after filtering: {len(seg_true)} - {round(100*len(seg_true)/len(seg),2)}%')
        print('=================================')

    if plot_confidence_distr==True:
        fig_suptitle=' - '.join([panel_sample_name,bays_model_name])

        fig,ax=plt.subplots(1,2,figsize=(6,3.5))
        sns.histplot()
        sns.histplot(seg['confidence'],bins=100,ax=ax[0])
        sns.histplot(seg['assignment_confidence'],bins=100,ax=ax[1])
        ax[0].axvline(confidence_thr,color='red')
        ax[1].axvline(assignment_confidence_thr,color='red')

        fig.suptitle(fig_suptitle,fontweight='bold',y=1)
        plt.tight_layout()


    return adata   


### =================================================================
## Create function that splits a range of numbers into equal sized bins + adjusts the last bin's size, as it can be different from bin_size
#  This function is used to create the coordinates of a 2D bins, binning the background signal of Baysor outputs in order to add the bins to an
#  andnata object 
def split_range_to_bins(start,end,bin_size):
    bins = np.arange(start, end + bin_size, bin_size)
    
    # If the last bin is larger than the specified bin size, adjust it
    if bins[-1] > end:
        bins[-1] = end
    
    return bins


### =================================================================
## Bin the background signal not assigned to cells into rectangles with side length of 'bin_rect_side_length' & create anndata
#  extract the gene counts for each rectangle-shaped bin into an anndata (each row corresponds to one bin, columns are genes)
def bin_background_signal_into_anndata(bg,bin_rect_side_length):
    import time
    import warnings
    warnings.filterwarnings("ignore")

    '''
    @bg: dataframe containing only transcripts that were not assigned to a cell and wee not considered as noise in the Baysor output
    @bin_rect_side_length: Background gets binned into squares with side length==bin_rect_side_length -> int
    '''
        
    # bin the data into equally spaced groups   
    x_bins=split_range_to_bins(bg['x'].min().astype(int),bg['x'].max().astype(int)+1,bin_rect_side_length+1)
    y_bins=split_range_to_bins(bg['y'].min().astype(int),bg['y'].max().astype(int)+1,bin_rect_side_length+1)
    bg.loc[:,'x_cut']=pd.cut(bg['x'], x_bins, right=False,include_lowest=False)
    bg.loc[:,'y_cut']=pd.cut(bg['y'], y_bins, right=False,include_lowest=False)

    ## Extract names and counts of genes for each square bin of the background
    a=bg[['x','y','x_cut','y_cut','gene']].groupby(['x_cut', 'y_cut']).apply(lambda x: x['gene'].value_counts()).reset_index().set_index(['x_cut', 'y_cut'])

    
    ## Extract gene names and counts for each bin and transform it into an expression matrix format (count_df)
    l=[]
    for gene,bg_bin in a.reset_index().set_index('gene').groupby('gene'):
        count_series=bg_bin.reset_index().set_index(['x_cut','y_cut']).drop(columns=['gene'])
        count_series=count_series.rename(columns={'count': gene})
        l.append(count_series)
    
    count_df=pd.concat(l,axis=1,join='outer').fillna(0)

    
    ## Create adata.obs dataframe
    obs=pd.DataFrame(index=np.arange(len(count_df.index)))
    
    # Create adata.var dataframe
    var=pd.DataFrame(index=count_df.columns)
    
    
    #### CREATE ANNDATA OBJECT
    # Extract the count matrix of true cells from count_df
    from scipy import sparse
    X=count_df.values
    adata=ad.AnnData(X, obs=obs, var=var, dtype='float64')
    adata.X=sparse.csr_matrix(adata.X)

    # Calculate some basic statistics   
    adata.obs['n_counts']=adata.X.sum(axis=1).A1
    adata.obs['n_genes']=(adata.X>0).sum(axis=1).A1
    adata.var['n_counts']=adata.X.sum(axis=0).A1
    adata.var['n_cells']=(adata.X>0).sum(axis=0).A1


    ## Extract x,y coordinates of bins and save them as an np.array for plotting the bins later
    bin_coords=a.index.drop_duplicates()
    
    x_coords=pd.IntervalIndex(bin_coords.get_level_values('x_cut'))
    y_coords=pd.IntervalIndex(bin_coords.get_level_values('y_cut'))
    
    x_coords_arr=np.transpose(np.array([x_coords.left.tolist(),x_coords.left.tolist(),(x_coords.right-1).tolist(),(x_coords.right-1).tolist()]))
    y_coords_arr=np.transpose(np.array([y_coords.left.tolist(),(y_coords.right-1).tolist(),y_coords.left.tolist(),(y_coords.right-1).tolist()]))
    polygon_vertex_coords=np.transpose(np.stack((x_coords_arr,y_coords_arr)),(1,2,0))
    adata.uns['polygon_vertex_coords']=polygon_vertex_coords


    ## Calculate and add centroids of bins to adata.obs + 
    adata.obs['x_centroid']=((x_coords.left+x_coords.right-1)/2).values
    adata.obs['y_centroid']=((y_coords.left+y_coords.right-1)/2).values
    adata.obsm["spatial"] = adata.obs[["x_centroid", "y_centroid"]].copy().to_numpy()

    return adata


### =================================================================
## Calculate the AUC values of the KDE smoothed avg_assignment_confidence curves:
#  - for cells with LOW average confidence (avg_assignment_confidence <=0.5)
#  - for cells with HIGH average confidence (avg_assignment_confidence >0.5)
#  - weighted: cells higher avg_assignment_confidence are more desirable to include in analysis 
#    -> weight the AUC of the KDE curve with the avg_assignment_confidence values (areas with higher avg_assignment_confidence values have higher weights)
def calculate_aucs(g,n,x_thr,weighted):
    '''
    g: sns.kdeplot to extract the KDE curve from
    n: number of curve on the KDEplot 
    x_thr: avg_assignment_confidence threshold to split low/high confidence cells
    weighted: boolean (True/False) if to use custom weighted trapezoidal AUC calcuation 
    '''
    c=g.get_lines()[n].get_data()
    x_val=c[0]
    y_val=c[1]

    ## weighted trapezoidal AUC calcuation (weighting by the height of each trapezoid(==the avg_assignment_confidence value of that bin))
    def weighted_trapz(y_val,x_val):
        trap_heights=(y_val[:-1] + np.diff(y_val)/2)
        delta_x=np.diff(x_val)
        trap_areas=trap_heights * delta_x
        
        ## Weight each trapezoids area with its height and then divide by the sum of the heights to normalize
        weighted_areas=(trap_areas*trap_heights)#/np.sum(trap_heights)
        summed_weighted_area=np.sum(weighted_areas)

        return summed_weighted_area
        
    ## Calculate AUC of avg_assignment_confidence KDE curve BELOW threshold  value
    idx=(x_val<=x_thr)
    sel_region_x=x_val[idx]
    sel_region_y=y_val[idx]
    
    if weighted==True:
        auc_all=weighted_trapz(y_val,x_val)
        auc_low=weighted_trapz(sel_region_y,sel_region_x)
        
        ## Standardise so that AUC adds to 1
        auc_low=auc_low/auc_all
    else:        
        auc_low=np.trapz(sel_region_y,sel_region_x)
    
    #g.fill_between (sel_region_x, sel_region_y, 0, facecolor='none', edgecolor='r', hatch='xx')

    
    ## Calculate AUC of avg_assignment_confidence KDE curve OVER threshold  value
    idx=(x_val>x_thr)
    sel_region_x=x_val[idx]
    sel_region_y=y_val[idx]
    
    if weighted==True:
        auc_high=weighted_trapz(sel_region_y,sel_region_x)
        ## Standardise so that AUC adds to 1
        auc_high=auc_high/auc_all
    else:    
        auc_high=np.trapz(sel_region_y,sel_region_x)

    #g.fill_between(sel_region_x, sel_region_y, 0, facecolor='none', edgecolor='b', hatch='xx')
    return auc_low,auc_high


### =================================================================
## Calculate Pearson correlation matrix between the expression vectors of igh quality cells and the expression vector of their summed neighbourhood
def calculate_corr_with_neighbouring_bg(adata,n_neighbors,max_radius,coord_unit='pixels',consider_low_quality_cells_as_background=False):
    '''
    adata: anndata containing the background bins and the foreground (cells)
    n_neighbors: number of neigbours (==k) to consider when building kNN neighbourhood => set to gigher number in order to catch a decent number of background bins
    max_radius: only consider kNN-background neighbours that are within the distance of maximal radius to the given cell, unit: PIXELS!!!
    consider_low_quality_cells_as_background: boolean, whether to consider low quality cells as background => 
                                              if True, they will also be treated as background and therefore \
                                              will be used to construct the kNN background neighbourhood
    '''

    from sklearn.neighbors import NearestNeighbors
    import logging
    logging.basicConfig(format='%(asctime)s - %(message)s', datefmt='%d-%b-%y %H:%M:%S')
    logging.getLogger().setLevel(logging.INFO)
    from scipy import sparse

    logging.info(f"Calculating correlation of cells with surrounding background bins. Parameters: n_neighbors:{n_neighbors}, max_radius:{max_radius}")
    
    ##======== 1.
    ## Extract coordinates of high_quality_cells 
    cell_coords=adata.obs.loc[adata.obs['label_for_knn_search']=='high_qual_cell',['x','y']].values
    
    if consider_low_quality_cells_as_background==True:
        bg_cats=['background','low_qual_cell']
    else:
        bg_cats=['background']   
    bg_coords=adata.obs.loc[adata.obs['label_for_knn_search'].isin(bg_cats),['x','y']].values 
    
    ##======== 2.
    ## Fit a NearestNeighbors model on array 'b'
    nbrs=NearestNeighbors(n_neighbors=n_neighbors, algorithm='auto').fit(bg_coords)
    
    ## Find the k-nearest neighbors for each point in array 'a'
    distances,indices=nbrs.kneighbors(cell_coords)

    ##======== 3.
    ## Mask those k-neares neigbours that are outside of the max_radius distance
    nbrs_inside_radius_mask=distances<=max_radius

    # Find the row indices with only False values ==> there were no neigbours within max_radius range
    indices_with_no_neighbors=np.where(np.all(nbrs_inside_radius_mask == False, axis=1))[0]
    logging.info(f"Number of high quality cells with no neighbours within {max_radius} {coord_unit} distance: {len(indices_with_no_neighbors)}")

    
    ##======== 4.
    ## For each cell sum the expression of the k-nearest neighbor and save it into a np.array 
    ngbr_expressions=np.array([c.X[indices[n][nbrs_inside_radius_mask[n]]].sum(axis=0) for n in range(indices.shape[0])])
    ngbr_expressions=np.squeeze(ngbr_expressions,axis=1)

    ##### DEPRECATED: CALCULATING THE MEAN EXPRESSION OF KNN NAIGHBOURHOOD DOESN'T CHANGE THE PEARSON CORRELATION, AS IT IS ONLY THE SUM/NUM_OF_BINS =>
    #####             DIVIDING BY SCALAR HAS NO EFFECT ON PEARSON CORRELATION
    ## Extract the number of NN background bins for each cell and use it to calculate the mean expression of the NN-expression
    #num_of_bg_bins_inside_radius=nbrs_inside_radius_mask.sum(axis=1)
    #ngbr_expressions_mean=ngbr_expressions/num_of_bg_bins_inside_radius[:, np.newaxis]
    #ngbr_expressions=ngbr_expressions_mean.copy()
    
    ## Calculate Pearson correlation matrix between the expression vectors for each high quality cell and the expression vector of its summed neighbourhood
    corr_mat=np.corrcoef(adata.X[adata.obs['label_for_knn_search']=='high_qual_cell'].A,ngbr_expressions,rowvar=True)
    
    #  - np.corrcoef() calculates the correlation between all possible pairs of a the 2 2D-matrices 
    #  => extract only the Pearson correlation between high quality cells & its summed neighbourhoods
    cell_corr_with_neighboring_bg=corr_mat[:cell_coords.shape[0],cell_coords.shape[0]:].diagonal()

    ## For cells where no neighbours were found in given radius, set the correlation to np.nan
    #cell_corr_with_neighboring_bg=cell_corr_with_neighboring_bg.copy()
    #cell_corr_with_neighboring_bg[indices_with_no_neighbors]=np.nan

    ##======== 5.
    ## Add Pearson correlations + to adata.obs
    adata.obs['Pearson_corr_with_bg_radius_'+str(max_radius)]=np.nan
    adata.obs.loc[adata.obs['label_for_knn_search']=='high_qual_cell','Pearson_corr_with_bg_radius_'+str(max_radius)]=cell_corr_with_neighboring_bg
    
    #adata.uns[f'{n_neighbors}_NN-neighbours_max_radius_{max_radius}_dict']={'distances':s_distances,
    #                                                                        'indices':indices,
    #                                                                        'indices_with_no_neighbors':indices_with_no_neighbors}

    adata.uns[f'{n_neighbors}_NN-neighbours_max_radius_{max_radius}_distances']=sparse.csr_matrix(distances)
    adata.uns[f'{n_neighbors}_NN-neighbours_max_radius_{max_radius}_indices']=indices
    adata.uns[f'{n_neighbors}_NN-neighbours_max_radius_{max_radius}_indices_with_no_neighbors']=indices_with_no_neighbors


    return adata


### =================================================================
## Align the segmentation.csv files from the same models but different bayasor scale parameters, and try to extract
#  cells in the lowest scale-parameters model (scale_5), that correspond to cells in found with models with higher scale parameter
#  i.e. model_scale_5        |model_scale_10        |counts
#       ---------------------------------------------------    
#       model_scale_5_cell_1 |model_scale_10_cell_10| 4    => model_scale_5_cell_1's transcripts were assigned to model_scale_10_cell_10 &
#       model_scale_5_cell_1 |model_scale_10_cell_19| 5       model_scale_10_cell_19 by model_scale_10, with counts 4,5 respectively.
#       model_scale_5_cell_2 |model_scale_10_cell_45| 35   => scale_5_cell_2's transcripts were all assigned to scale_10_cell_45 by model_scale_10
#       model_scale_5_cell_3 |model_scale_10_cell_45| 5    => scale_5_cell_3's transcripts were also assigned to scale_10_cell_45
#       model_scale_5_cell_4 |model_scale_10_cell_22| 67   => scale_5_cell_4's transcripts were exclusively assigned to scale_10_cell_22
#       model_scale_5_cell_5 |NaN                   |      => scale_5_cell_5's transcipts weren't assigned to a cell by model_scale_10
##   
## This way one can have an estimate which cell corresponds to which one between the models. Typically, cells from lower scale are merged
#  to one bigger cell when using a model with higher scale
def count_overlapping_transcripts_between_matching_cells(transcript_cell_assign_df,bays_model_to_compare_to,transcr_belonging_to_bays_model_name_min_scale):
    for col in transcript_cell_assign_df.columns:
        if col !='cell_'+bays_model_to_compare_to:
            transcr_belonging_to_bays_model_name_min_scale[col]=transcript_cell_assign_df.groupby('cell_'+bays_model_to_compare_to,dropna=False).apply(lambda x:x[col].value_counts(dropna=False))      
    
    return transcr_belonging_to_bays_model_name_min_scale  


### =================================================================
## Split cells into low quality/high quality cells based on average_assignment_confidence and n_count thresholds
#  and add these labels as a new column ('label_for_knn_search')      
def define_low_high_quality_cells(c,avg_asg_thr,n_counts_quant_thr):
    #print(adata.obs['n_counts'].quantile(quant_thr))
    n_counts_thr=c.obs.loc[c.obs['origin']=='cell','n_counts'].quantile(n_counts_quant_thr)

    c.obs['label_for_knn_search']='high_qual_cell'

    c.obs.loc[(c.obs['origin']=='cell')\
              &((c.obs['avg_assignment_confidence']<=avg_asg_thr)\
                |(c.obs['n_counts']<=n_counts_thr))\
                 ,'label_for_knn_search']='low_qual_cell'

    c.obs.loc[c.obs['origin']=='background','label_for_knn_search']='background' 

    return c


### =================================================================
## Calculate Pearson correlation matrix between the expression vectors of high quality cells and the expression vector of their summed neighbourhood
def return_neighbouring_bg_expr(adata,n_neighbors,max_radius,consider_low_quality_cells_as_background,coord_unit='pixels'):
    '''
    adata: anndata containing the background bins and the foreground (cells)
    n_neighbors: number of neigbours (==k) to consider when building kNN neighbourhood => set to gigher number in order to catch a decent number of background bins
    max_radius: only consider kNN-background neighbours that are within the distance of maximal radius to the given cell, unit: PIXELS!!!
    consider_low_quality_cells_as_background: boolean, whether to consider low quality cells as background => 
                                              if True, they will also be treated as background and therefore \
                                              will be used to construct the kNN background neighbourhood
    '''

    from sklearn.neighbors import NearestNeighbors
    import logging
    logging.basicConfig(format='%(asctime)s - %(message)s', datefmt='%d-%b-%y %H:%M:%S')
    logging.getLogger().setLevel(logging.INFO)
    from scipy import sparse

    logging.info(f"Calculating correlation of cells with surrounding background bins. Parameters: n_neighbors:{n_neighbors}, max_radius:{max_radius}")
    
    ##======== 1.
    ## Extract coordinates of high_quality_cells 
    cell_coords=adata.obs.loc[adata.obs['label_for_knn_search']=='high_qual_cell',['x','y']].values
    
    if consider_low_quality_cells_as_background==True:
        bg_cats=['background','low_qual_cell']
    else:
        bg_cats=['background']   
    bg_coords=adata.obs.loc[adata.obs['label_for_knn_search'].isin(bg_cats),['x','y']].values 

    ##======== 2.
    ## Fit a NearestNeighbors model on array 'b'
    nbrs=NearestNeighbors(n_neighbors=n_neighbors, algorithm='auto').fit(bg_coords)
    
    ## Find the k-nearest neighbors for each point in array 'a'
    distances,indices=nbrs.kneighbors(cell_coords)

    ##======== 3.
    ## Mask those k-neares neigbours that are outside of the max_radius distance
    nbrs_inside_radius_mask=distances<=max_radius

    # Find the row indices with only False values ==> there were no neigbours within max_radius range
    indices_with_no_neighbors=np.where(np.all(nbrs_inside_radius_mask == False, axis=1))[0]
    logging.info(f"Number of high quality cells with no neighbours within {max_radius} {coord_unit} distance: {len(indices_with_no_neighbors)}")

    
    ##======== 4.
    ## For each cell sum the expression of the k-nearest neighbor and save it into a np.array 
    ngbr_expressions=np.array([c.X[indices[n][nbrs_inside_radius_mask[n]]].sum(axis=0) for n in range(indices.shape[0])])
    ngbr_expressions=np.squeeze(ngbr_expressions,axis=1)

    return ngbr_expressions


### =================================================================
### 1. Input: Cells produced by Baysor_model_X_scale_Y (==base) matched up with cells produced by Baysor_model_X_scale_Z (==compare_to) 
### 2. Calculate difference between expression of matching cells from the two methods (as they transcripts not necessarily completely overlap)
### 3. Correlate the expression difference to the neighbouring background of base method 
###    ==> if high correlation with background, the compared_to method matching cell contains mostly backround noise compared to cell of base method
###    ==> in these cases, compared_to method's scale parameter might be too large (?)
def calc_corr_between_neigh_bg_and_non_overlapping_transcr_from_matching_cells(adata_base,adata_to_compare,
                                                                           one_to_one_cells,
                                                                           bays_model_to_compare_to,
                                                                           bays_model_to_compare,
                                                                           n_neighbors,
                                                                           max_radius_list,
                                                                           consider_low_quality_cells_as_background=False):

    ## Subset adata from base model to cells 
    hq_cells_base=adata_base.obs[adata_base.obs['label_for_knn_search']=='high_qual_cell'].index.tolist()
    one_to_one_cells_base=one_to_one_cells['cell_'+bays_model_to_compare_to].tolist()
    cell_ids_to_drop=list(set(hq_cells_base) - set(one_to_one_cells_base))
    
    adata_base_cells_hq=adata_base[#(adata_base.obs['label_for_knn_search']=='high_qual_cell')\
                                   ~(adata_base.obs.index.isin(cell_ids_to_drop)),:].copy()

    ## Subset matching cell IDs to contain only high-qaulity cells from base method
    one_to_one_cells_hq=one_to_one_cells[(one_to_one_cells['cell_'+bays_model_to_compare_to].isin(hq_cells_base))]
    print(f'Number of one_to_one_cells_hq: {one_to_one_cells_hq.shape[0]}')
    
    
    ## Subset adata from model to compare to cells   
    adata_to_compare_cells=adata_to_compare[adata_to_compare.obs['origin']=='cell',:].copy()

    ## For each cell give a unique number, which will be the corresponding rownumber of the cell in the expression matrix
    d1=dict(zip(adata_base_cells_hq.obs.index.tolist(),np.arange(len(adata_base_cells_hq.obs.index))))
    d2=dict(zip(adata_to_compare_cells.obs.index.tolist(),np.arange(len(adata_to_compare_cells.obs.index))))
    
    cell_indices_base=one_to_one_cells_hq['cell_'+bays_model_to_compare_to].map(d1).tolist()
    cell_indices_to_compare=one_to_one_cells_hq[bays_model_to_compare].map(d2).tolist()
    
    expr_adata_base_cells=adata_base_cells_hq.layers['raw_counts'].A
    expr_adata_to_compare_cells=adata_to_compare_cells.layers['raw_counts'].A

    ## Calculate the difference between the matching cells produced by the two methods 
    #  => as most of the matching cells don't overlap completely, the difference between their expressions can be compared to the neighbouring background
    #  => if the difference correlates with the neighbouring background, the matching cell produced by the to-be-compared scale method added mostly noise to the cell
    #     produced by the base method => ====== to-be-compared method's scale parameter is too large?  =========
    non_overlapping_transc_between_models=np.abs(expr_adata_base_cells[cell_indices_base] - expr_adata_to_compare_cells[cell_indices_to_compare])
    
    for max_radius in max_radius_list:
        ngbr_expressions=return_neighbouring_bg_expr(adata_base_cells_hq,n_neighbors,max_radius,consider_low_quality_cells_as_background,coord_unit='pixels')
         
        corr_mat=np.corrcoef(non_overlapping_transc_between_models,ngbr_expressions,rowvar=True)
        corr_non_overlapping_transc_with_bg=corr_mat[:one_to_one_cells_hq.shape[0],one_to_one_cells_hq.shape[0]:].diagonal()

        obs_corr_coln=bays_model_to_compare_to.split('cell_')[-1]+'_corr_with_'+bays_model_to_compare

        scale_size=bays_model_to_compare.split('scale')[-1].split('_')[1]
        corr_coln='corr_bw_bg_and_non_overlapping_transc_with_scale_'+scale_size+'_radius_'+str(max_radius)+'_n_neigh_'+str(n_neighbors)
        adata_base.obs[corr_coln]=np.nan
        adata_base.obs.loc[one_to_one_cells_hq['cell_'+bays_model_to_compare_to].tolist(),corr_coln]=corr_non_overlapping_transc_with_bg

    
    return adata_base



### =================================================================
## Calculate correlation of base method cells with matching cells produced by the methods we want to compare base method to
#  - Matching cells: for each cell produced by base method, a matching cell from extract 
### 1. Input: Cells produced by Baysor_model_X_scale_Y (==base) matched up with cells produced by Baysor_model_X_scale_Z (==compare_to) 
###           - one_to_one cells: each cell's transcripts from base method were assigned to one cell from compare_to method 
###           - Possible, that multiple cells from base method were assigned to the same compare_to cell (==they were merged into one bigger cell)
### 2. Correlate the expression of matching cells from the two methods (as they transcripts mostly don't completely overlap)
###    ==> if low correlation seen in matching cells, the compared_to method matching cell is different from the cell of base method
###    ==> in these cases, compared_to method's scale parameter merged base cells, that are different --> scale might be too large (?)
###    ==> in these cases, compared_to method's scale parameter merged base cells, that are subcompartments of the same bigger cell --> scale might be too small (?)

def calculate_corr_between_matching_cells(adata_base,adata_to_compare,one_to_one_cells,bays_model_to_compare_to,bays_model_to_compare):
    ## Subset adata from base model to cells and background 
    hq_cells_base=adata_base.obs[adata_base.obs['label_for_knn_search']=='high_qual_cell'].index.tolist()
    one_to_one_cells_base=one_to_one_cells['cell_'+bays_model_to_compare_to].tolist()
    cell_ids_to_drop=list(set(hq_cells_base) - set(one_to_one_cells_base))
    
    adata_base_cells_hq=adata_base[#(adata_base.obs['label_for_knn_search']=='high_qual_cell')\
                                   ~(adata_base.obs.index.isin(cell_ids_to_drop)),:].copy()
                       
    one_to_one_cells_hq=one_to_one_cells[(one_to_one_cells['cell_'+bays_model_to_compare_to].isin(hq_cells_base))]
        
    
    ## Subset adata from model to compare to cells   
    adata_to_compare_cells=adata_to_compare[adata_to_compare.obs['origin']=='cell',:].copy()
    

    ## For each cell give a unique number, which will be the corresponding rownumber of the cell in the expression matrix
    d1=dict(zip(adata_base_cells_hq.obs.index.tolist(),np.arange(len(adata_base_cells_hq.obs.index))))
    d2=dict(zip(adata_to_compare_cells.obs.index.tolist(),np.arange(len(adata_to_compare_cells.obs.index))))
    
    cell_indices_base=one_to_one_cells_hq['cell_'+bays_model_to_compare_to].map(d1).tolist()
    cell_indices_to_compare=one_to_one_cells_hq[bays_model_to_compare].map(d2).tolist()
    
    expr_adata_base_cells=adata_base_cells_hq.layers['raw_counts'].A
    expr_adata_to_compare_cells=adata_to_compare_cells.layers['raw_counts'].A
    

    ## Calculate correlation of base method cells with matching cells produced by the methods we want to compare base method to
    corr_mat=np.corrcoef(expr_adata_base_cells[cell_indices_base],expr_adata_to_compare_cells[cell_indices_to_compare],rowvar=True)
    cell_corr_with_matching_cells=corr_mat[:one_to_one_cells_hq.shape[0],one_to_one_cells_hq.shape[0]:].diagonal()

    corr_coln=bays_model_to_compare_to.split('cell_')[-1]+'_corr_with_'+bays_model_to_compare
    adata_base.obs[corr_coln]=np.nan
    adata_base.obs.loc[one_to_one_cells_hq['cell_'+bays_model_to_compare_to].tolist(),corr_coln]=cell_corr_with_matching_cells

    one_to_one_cells_hq['corr']=cell_corr_with_matching_cells

    return adata_base,one_to_one_cells_hq



####=======================
## Function to extract the proper correlations from a large np.corrcoef() matrix + calculate the mean correlation within cell groups
def calculate_mean_corr_within_cell_groups(array, split_indices):
    corr_within_cell_groups=[]
    start_row,start_col=0,0
    
    for idx in split_indices:
        end_row = start_row + idx
        end_col = start_col + idx

        split=array[start_row:end_row, start_col:end_col]
        ## Extract correlations of cells with each other that are merged into one bigger cell (1 element above the digonal)
        upper_triangle=split[np.triu_indices(split.shape[0],k=1)]
        
        corr_upper_triangle=np.mean(upper_triangle)
        
        corr_within_cell_groups.append(corr_upper_triangle)
        start_row,start_col=end_row, end_col
    
    return corr_within_cell_groups



## Cells from smaller scale value methods tend to be smaller, and methods with higher scale values can merge these smaller cells into one larger cell.
    ## Calculate the correlation within the base method cells that get merged into a bigger cell produced by the methods we want to compare base method to
    #  i.e. Baysor_model_X_scale_Y_Cell_1,2,3 are merged into one cell by Baysor_model_X_scale_Z, then compute correlation within Baysor_model_X_scale_Y_Cell_1,2,3
    
    #  ==> if correlation is high, this could mean that Baysor_model_X_scale_Z is merging similar cells into doublets ? ==> scale_Z too large (?)
    #  ==> if correlation is low, this could mean that Baysor_model_X_scale_Z is merging dissimlar cells into doublets ? ==> scale_Z too large (?)
    #  ==> if correlation is low, this could mean that Baysor_model_X_scale_Z is merging subcellular compartments into cells ? ==> scale_Z is okay (?)
def calculate_corr_between_smaller_cells_merged_to_a_larger(adata_base,adata_to_compare,all_cells_df,bays_model_to_compare_to,bays_model_to_compare):
    
    ## Subset adata from base model to cells and background 
    hq_cells_base=adata_base.obs[adata_base.obs['label_for_knn_search']=='high_qual_cell'].index.tolist()
    all_cells_base=all_cells_df['cell_'+bays_model_to_compare_to].tolist()
    cell_ids_to_drop=list(set(hq_cells_base) - set(all_cells_base))
    
    adata_base_cells_hq=adata_base[#(adata_base.obs['label_for_knn_search']=='high_qual_cell')\
                                   ~(adata_base.obs.index.isin(cell_ids_to_drop)),:].copy()

    ## Sort values tso that the cell indices align between the two methods after calculation of 'indices_of_cell_groupings' below
    all_cells_df_hq=all_cells_df[(all_cells_df['cell_'+bays_model_to_compare_to].isin(hq_cells_base))].dropna().sort_values(by=bays_model_to_compare)

    
    ## Subset adata from model to compare to cells   
    adata_to_compare_cells=adata_to_compare[adata_to_compare.obs['origin']=='cell',:].copy()
    

    ## For each cell give a unique number, which will be the corresponding rownumber of the cell in the expression matrix
    d1=dict(zip(adata_base_cells_hq.obs.index.tolist(),np.arange(len(adata_base_cells_hq.obs.index))))
    d2=dict(zip(adata_to_compare_cells.obs.index.tolist(),np.arange(len(adata_to_compare_cells.obs.index))))
    
    cell_indices_base=all_cells_df_hq['cell_'+bays_model_to_compare_to].map(d1).tolist()
    cell_indices_to_compare=all_cells_df_hq[bays_model_to_compare].map(d2).tolist()
    
    expr_adata_base_cells=adata_base_cells_hq.layers['raw_counts'].A
    expr_adata_to_compare_cells=adata_to_compare_cells.layers['raw_counts'].A
    

    ## Calcaute correlation matrix between the smaller scale cells and extract the mean correlations within the cell groups that are 
    #  merged by the larger scale method
    corr_mat=np.corrcoef(expr_adata_base_cells[cell_indices_base],rowvar=True)
    indices_of_cell_groupings=all_cells_df_hq.groupby(bays_model_to_compare,dropna=True).apply(lambda x: len(x)).values
    corr_within_cell_groups=calculate_mean_corr_within_cell_groups(corr_mat,indices_of_cell_groupings)
    #print(len(corr_within_cell_groups))
    #print(all_cells_df_hq.groupby(bays_model_to_compare,dropna=False).apply(lambda x: len(x)).index.tolist())

    corr_coln='mean_corr_within_cells_merged_cells_from_'+bays_model_to_compare_to.split('cell_')[-1]
    adata_to_compare.obs[corr_coln]=np.nan
    adata_to_compare_cell_ids=all_cells_df_hq.groupby(bays_model_to_compare,dropna=False).apply(lambda x: len(x)).index.tolist()
    adata_to_compare.obs.loc[adata_to_compare_cell_ids,corr_coln]=corr_within_cell_groups

    #one_to_one_cells_hq['corr']=cell_corr_with_matching_cells
    
    return adata_to_compare
```

# Match leiden clusters between the methods

## Merge adatas with different scales and check compare their leiden clusters

```python
from scipy import sparse

medoid_dict={}
adata_hq_merge_dict={}
for presegmentation_method in presegmentation_methods[:]:
    
    scale_size=5 # can be 5,10,15
    bays_model_to_compare_to=[meth_name for meth_name in adata_dict[presegmentation_method].keys() if 'scale_'+str(scale_size) in meth_name][0]

    l=[]
    adata_hq_list=[]
    medoid_dict[presegmentation_method]={}
    
    #fig,ax=plt.subplots(1,1,figsize=(16,4))
    for bays_model,adata_bays_model in adata_dict[presegmentation_method].items():#[bays_model_to_compare_to]
        print(bays_model)
        ## Define high quality cells
        avg_asg_thr=0.5
        n_counts_quant_thr=0.2
        adata_bays_model=define_low_high_quality_cells(adata_bays_model,avg_asg_thr,n_counts_quant_thr)
        adata_hq=adata_bays_model[adata_bays_model.obs['label_for_knn_search']=='high_qual_cell',:].copy()
        #sc.pp.log1p(adata_base_hq)
        adata_hq.layers['norm_by_area']=sparse.csr_matrix(adata_hq.layers['raw_counts'].multiply(1/(adata_hq.obs['area'].values.reshape(-1,1))))

        adata_hq.obsm["norm_by_area_pca"] = sc.tl.pca(adata_hq.layers["norm_by_area"],n_comps=20)
        sc.pp.neighbors(adata_hq,use_rep='norm_by_area_pca',n_pcs=20,metric='euclidean')
        #sc.tl.umap(adata_hq,random_state=0)
        sc.tl.leiden(adata_hq,resolution=0.3,key_added='leiden')
    
        adata_hq_list.append(adata_hq)

        ## Calculate the medoids of the leiden clusters using PCA
        pca_repr=pd.DataFrame(index=adata_hq.obs['leiden'],data=adata_hq.layers['norm_by_area'].A)  #adata_hq.obsm["norm_by_area_pca"]
        leiden_medoids_pca=pca_repr.groupby('leiden').apply(lambda x: x.mean(axis=0))
        medoid_dict[presegmentation_method][bays_model]=leiden_medoids_pca
        
        print('Number of leiden clusters: ',len(adata_hq.obs['leiden'].unique()),'\n')

        df=adata_hq.obs#[['area','leiden']]
        df['model']=bays_model
        l.append(df)

    adata_hq_merge=ad.concat(adata_hq_list,axis=0,merge='same',label='model',keys=[*adata_dict[presegmentation_method]])
    adata_hq_merge_dict[presegmentation_method]=adata_hq_merge
    

           
```

## Clustering within scale-parameter model

### Cluster medoids comparison - Correct for possible batch effects arising from different scale parameters

```python
for presegmentation_method in presegmentation_methods[:]:
    print(adata_hq_merge)

    adata_hq_merge.X=adata_hq_merge.layers['norm_by_area']
    sc.pp.combat(adata_hq_merge, key='model')
    
    expr=pd.DataFrame(index=pd.MultiIndex.from_frame(adata_hq_merge.obs[['leiden','model']]),data=adata_hq_merge.X.A)  #adata_hq.obsm["norm_by_area_pca"]
    
    leiden_medoids_pca=expr.groupby(['leiden','model']).apply(lambda x: x.median(axis=0)).dropna()
    
    embedding = umap.UMAP(n_components=2, min_dist=0.001, metric='correlation',random_state=42, verbose=False).fit_transform(leiden_medoids_pca)
    umap_df=pd.DataFrame(index=leiden_medoids_pca.index,data=embedding,columns=['UMAP1','UMAP2'])
    umap_df['label']='scale_' + (umap_df.reset_index()['model'].str.split('_exp_',expand=True)[0]).str.split('_scale_',expand=True)[1].values\
                     +'\nleiden_'+umap_df.index.get_level_values('leiden').astype(str)
    
    fig,ax=plt.subplots(1,1,figsize=(12,8))
    fig.suptitle(presegmentation_method+' -  UMAP of leiden cluster expression medians',y=0.95)
    p1=sns.scatterplot(data=umap_df,x='UMAP1',y='UMAP2',hue='model',ax=ax,s=55,legend=True)
    
    for line in range(0,umap_df.shape[0]):
         p1.text(umap_df['UMAP1'].values[line]+0.01, umap_df['UMAP2'].values[line]+0.04, 
         umap_df['label'].values[line], horizontalalignment='center', 
         size='small', color='black')#, weight='semibold')
```

### Cluster medoids comparison - No batch effect correction between the models with different scale parameters

```python
from sklearn.neighbors import NearestNeighbors
import umap
for presegmentation_method in presegmentation_methods[:]:
    
    scale_size=5 # can be 5,10,15
    bays_model_to_compare_to=[meth_name for meth_name in adata_dict[presegmentation_method].keys() if 'scale_'+str(scale_size) in meth_name][0]

    base_medoids=medoid_dict[presegmentation_method][bays_model_to_compare_to]
    print(bays_model_to_compare_to,'model num of leiden clusters:',len(base_medoids))

    n_neighbors=1
    
    ## Fit a NearestNeighbors model on array 'b'
    l=[]

    ## Determine the nearest neighboring leiden clusters
    for bays_model,medoid in medoid_dict[presegmentation_method].items():
        medoid=medoid.reset_index()
        medoid['model']=bays_model
        medoid=medoid.set_index(['leiden','model'])
        l.append(medoid)
        
        if bays_model!=bays_model_to_compare_to:
            compare_medoids=medoid_dict[presegmentation_method][bays_model]
            print(bays_model,'model num of leiden clusters:',len(compare_medoids))
            
            ## Find the k-nearest neighbors for each clusters' medoid expression
            nbrs=NearestNeighbors(n_neighbors=n_neighbors, algorithm='auto').fit(compare_medoids.values)
            distances,indices=nbrs.kneighbors(base_medoids.values)
            #print(indices)
    print()
    ## PLot UMAP of expressions means
    adata_hq_merge=pd.concat(l,axis=0)        
    embedding = umap.UMAP(n_components=2, min_dist=0.001, metric='correlation',random_state=42, verbose=False).fit_transform(adata_hq_merge)
    umap_df=pd.DataFrame(index=adata_hq_merge.index,data=embedding,columns=['UMAP1','UMAP2'])
    umap_df['label']='scale_' + (umap_df.reset_index()['model'].str.split('_exp_',expand=True)[0]).str.split('_scale_',expand=True)[1].values\
                     +'\nleiden_'+umap_df.index.get_level_values('leiden').astype(str)
    
    fig,ax=plt.subplots(1,1,figsize=(12,8))
    fig.suptitle(presegmentation_method+' -  UMAP of leiden cluster expression means',y=0.95)
    p1=sns.scatterplot(data=umap_df,x='UMAP1',y='UMAP2',hue='model',ax=ax,s=55,legend=True)
    
    for line in range(0,umap_df.shape[0]):
         p1.text(umap_df['UMAP1'].values[line]+0.01, umap_df['UMAP2'].values[line]+0.04, 
         umap_df['label'].values[line], horizontalalignment='center', 
         size='small', color='black')#, weight='semibold')
```

```python
#fn='../../xenium_data/processed_data/true_cell_filtering/baysor/Panel1_P1_D/test_out.h5ad'
#adata_hq_merge_=sc.read_h5ad(fn)
#print(adata_hq_merge_)

def run_scDblFinder(adata,show_r_output):
    import tempfile
    import os
    import subprocess
    import pandas as pd
    import scanpy as sc
    with tempfile.TemporaryDirectory() as tmpdirname:
        ifn, ofn, rfn = [os.path.join(tmpdirname, e) for e in ["in.h5ad", "out.h5ad","script.R"]]
        adata.write_h5ad(ifn,compression='gzip')
        rcmd='.libPaths("/data/gpfs/projects/punim2121/R_libs/4.2.1");\
            library(zellkonverter);\
            library(scDblFinder);\
            library(SingleCellExperiment);\
            library(BiocParallel);\
            dat=zellkonverter::readH5AD("{0}",reader="python");\
            assayNames(dat)[grepl("raw_counts",assayNames(dat))]="counts";\
            bp<-MulticoreParam(workers=4, progressbar=FALSE,RNGseed=1234);\
            sce=scDblFinder(dat,samples="model",cluster="leiden",BPPARAM=bp);\
            assayNames(sce)[grepl("counts",assayNames(sce))]="raw_counts";\
            writeH5AD(sce,file="{1}",compression="gzip");'.format(ifn,ofn)
   
        #rcmd = rcmd.replace('\\', '\\\\')
        with open(rfn, "w") as f:
            f.write(rcmd)
        print("Running Doublet finder normalization...")
        #proc = subprocess.Popen(["Rscript", rfn,], stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
        proc=subprocess.Popen(f"module load foss/2022a R/4.2.1;R < {rfn} --no-save",
                              shell=True,stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
        while not proc.poll():
            ## To print output of R script
            output = proc.stdout.readline()
            if show_r_output==True:
                if output:
                    try:
                        print(output.decode("utf-8"))
                    except UnicodeDecodeError as e:
                        print(f"Error decoding output: {e}")
                        print("Raw output:", output)   
                c = proc.stdout.read(1)
            if not c:
                break
        #print("Saving acosh dataframe")
        adata_with_doublets=sc.read_h5ad(ofn)
    
    return adata_with_doublets


presegmentation_methods=['10x','no_segmentation','CPc','CPn']
for presegmentation_method in presegmentation_methods[:]:
    #scale_size=10 # can be 5,10,15
    #bays_model_to_compare_to=[meth_name for meth_name in adata_dict[presegmentation_method].keys() if 'scale_'+str(scale_size) in meth_name][0]

    adata_hq_merge=adata_hq_merge_dict[presegmentation_method]

    adata_hq_merge=run_scDblFinder(adata_hq_merge,show_r_output=False)
    adata_hq_merge_dict[presegmentation_method]=adata_hq_merge
        

#fig,ax=plt.subplots(1,1,figsize=(16,4))
#sns.boxplot(data=adata_hq_merge_.obs,x='leiden',y='scDblFinder.score',hue='model',ax=ax,order=adata_hq_merge.obs['leiden'].astype(int).sort_values().astype(str))
#ax.set_yscale('log')
```

### Calculate distance from expression means of Leiden clusters

```python

import scipy

leid_model_cell_freqs=adata_hq_merge.obs.groupby('leiden').apply(lambda x: (x['model'].value_counts())).reset_index()

adata_hq_merge.obs['dist_from_leid_clust_mean']=np.nan

for leid,leid_df in adata_hq_merge.obs.groupby('leiden'):

    means=[]   
    for bays_model,leid_model_df in leid_df.groupby('model'):
        
        cell_freq=leid_model_cell_freqs.loc[(leid_model_cell_freqs['leiden']==leid)&(leid_model_cell_freqs['model']==bays_model),'count']
        expr=pd.DataFrame(index=pd.MultiIndex.from_frame(leid_model_df[['leiden','model']]),
                          data=adata_hq_merge[(adata_hq_merge.obs['leiden']==leid)&\
                                              (adata_hq_merge.obs['model']==bays_model)].layers["norm_by_area"].A)  #adata_hq.obsm["norm_by_area_pca"]
        
        leiden_model_mean_expr=expr.groupby(['leiden','model']).apply(lambda x: x.mean(axis=0)).dropna()
        weight_leiden_model_mean_expr=(1/cell_freq.values)*leiden_model_mean_expr.values
        #means.append(weight_leiden_model_mean_expr)
        means.append(leiden_model_mean_expr.values)

    weight_mean_denom=(1/leid_model_cell_freqs.loc[(leid_model_cell_freqs['leiden']==leid),'count'].values).sum()
    #leid_clust_weight_expr_mean=np.sum(means,axis=0)/weight_mean_denom
    leid_clust_weight_expr_mean=np.mean(means,axis=0)

    expr_leid_clust=adata_hq_merge[adata_hq_merge.obs['leiden']==leid].layers["norm_by_area"].A
    dist = scipy.spatial.distance.cdist(expr_leid_clust,leid_clust_weight_expr_mean,metric='cosine')

    adata_hq_merge.obs.loc[adata_hq_merge.obs['leiden']==leid,'dist_from_leid_clust_mean']=dist

fig,ax=plt.subplots(1,1,figsize=(16,4))
sns.boxplot(data=adata_hq_merge.obs,x='leiden',y='dist_from_leid_clust_mean',hue='model',ax=ax)#,order=adata_hq_merge.obs['leiden'].astype(int).sort_values().astype(str))
#ax.set_yscale('log')
```
