# Defining functions

```python
## Setup function for calculating elapsed time
def print_elapsed_time(start,stop):
    # Calculate the elapsed time in seconds
    elapsed_seconds=stop-start
    
    # Convert elapsed time to hours and minutes
    elapsed_minutes, elapsed_seconds = divmod(int(elapsed_seconds), 60)
    elapsed_hours, elapsed_minutes = divmod(elapsed_minutes, 60)
    
    # Print the result in the desired format
    return(f"{elapsed_hours} hours: {elapsed_minutes} minutes: {elapsed_seconds} seconds") 

### FILTER low confidence transcripts & create anndata from Baysor output
def generate_anndata(panel_sample_name,bays_model_name,seg,seg_stats,nucleus_polygon_pixel_metrics,confidence_thr=0.8,
                     assignment_confidence_thr=0.5,qv_thr=20,print_filtering_results=True,plot_confidence_distr=False):

    import logging
    logging.basicConfig(format='%(asctime)s - %(message)s', datefmt='%d-%b-%y %H:%M:%S')
    logging.getLogger().setLevel(logging.INFO)

    ### Columns names of 'segmentation.csv' (as 'seg' df variable) dataframe => 
    #   Based on the description if the output columns found at https://github.com/kharchenkolab/Baysor
    '''
    confidence: probability of a molecule to be real (i.e. not noise)
    cell: id of the assigned cell. Value "" corresponds to noise.
    cluster: id of the molecule cluster
    assignment_confidence: confidence that the molecule is assigned to a correct cell
    is_noise: shows whether molecule was assigned to noise (it equals true if and only if cell == "")
    ncv_color: RGB code of the neighborhood composition coloring
    '''
    
    #### FILTER TRANSCRCIPTS WITH LOW CONFIDENCE
    ## Filter result of baysor segmentation based on the threshold values 
    seg_true=seg[(~seg['cell'].isna())&\
                (seg['confidence']>confidence_thr)&\
                (seg['assignment_confidence']>assignment_confidence_thr)&\
                (seg['qv']>qv_thr)] 


    #### EXTRACT COUNT MATRIX OF TRANSCRIPTS
    ## Extract the transcript counts for each gene for each cell 
    a=seg_true.groupby('cell').apply(lambda x: (x['gene'].value_counts(dropna=False)))

    l=[]
    for cell_series in a.index.get_level_values('cell').unique():
        # Rename 'count' column to gene => This will be the index column with the gene names in the final concatenated dataframe
        c=a[cell_series]
        c.rename(cell_series, inplace=True)
        l.append(c)

    ## Concatenate the list of pd.series of gene counts for each cell
    count_df=pd.concat(l,axis=1,join='outer').T.fillna(0)
    count_df=count_df.rename(index={'gene': 'cell_id'})


    #### CREATE ADATA.OBS & ADATA.VAR DATAFRAME
    ## Set cell_id as index for seg_stats df to make it uniform with other df-s
    seg_stats=seg_stats.rename(columns={'cell': 'cell_id'})
    seg_stats=seg_stats.set_index('cell_id')

    ## Extract list of True cell names => intersection of filtered cells (count_df) and 
    #  cells with non-zero polygon mask (nucleus_polygon_pixel_metrics contains only cells with non-zero masks see function 
    #  "create_cell_bound_df()" which created cell_bound_df)
    true_cells_ind=list(set(count_df.index)&set(nucleus_polygon_pixel_metrics.index))

    ## Add metrics of cell mask DAPI intensities and subset only to true cells 
    mask_stat_cols=['mean','median','max','mode']
    seg_stats_=seg_stats.copy()
    seg_stats_.loc[true_cells_ind,mask_stat_cols]=nucleus_polygon_pixel_metrics.loc[true_cells_ind,mask_stat_cols].values
    seg_stats_=seg_stats_.rename(columns={'mean':'mean_DAPI_intensity','median':'median_DAPI_intensity','max':'max_DAPI_intensity','mode':'mode_DAPI_intensity'})
    
    # Calculate log10 values of intensity metrics as they have a large range
    for metric in mask_stat_cols:
        colname='_'.join([metric,'DAPI_intensity'])
        log_colname='_'.join(['log',metric,'DAPI_intensity'])
        seg_stats_[log_colname]=np.log10(seg_stats_[colname])

    # Add sample and patient metadata
    seg_stats_['original_sample']=panel_sample_name
    seg_stats_['panel']=panel_sample_name.split('_')[0]
    seg_stats_['patient']=panel_sample_name.split('_')[1]
    seg_stats_['condition']=panel_sample_name.split('_')[-1]

    
    ## Calculate the area of the cell masks 
    #  => segmentation_stats.csv 'area' column: area of the convex hull around the cell molecules, not the cell masks' area
    cell_mask_areas=nucleus_polygon_pixel_metrics['raw_vals'].apply(lambda x: x.shape[0])
    seg_stats_['cell_mask_area']=cell_mask_areas[true_cells_ind]

   
    ## Create adata.obs dataframe
    obs=seg_stats_.loc[true_cells_ind,:].copy()

    # Create adata.var dataframe
    var=pd.DataFrame(index=count_df.columns)


    #### CREATE ANNDATA OBJECT
    # Extract the count matrix of true cells from count_df
    from scipy import sparse
    X=count_df.loc[true_cells_ind,:].values
    adata=ad.AnnData(X, obs=obs, var=var, dtype='float64')
    adata.X=sparse.csr_matrix(adata.X)
    adata.layers['raw_counts']=adata.X.copy()
    adata.layers['norm_by_area']=sparse.csr_matrix(adata.layers['raw_counts'].multiply(1/(adata.obs['cell_mask_area'].values.reshape(-1,1))))


    #### ADD ADDITIONAL METADATA & CALCULATE BASIC STATISTICS
    ## Add noise information
    pct_noise=sum(seg['cell'].isna())/len(seg)
    adata.uns['pct_noise']=pct_noise

    ## Add raw DAPI intensity values of cell masks
    #adata.uns['raw_DAPI_intensities']=nucleus_polygon_pixel_metrics['raw_vals']

    # Calculate some basic statistics   
    adata.obs['n_counts']=adata.X.sum(axis=1).A1
    adata.obs['n_genes']=(adata.X>0).sum(axis=1).A1
    adata.var['n_counts']=adata.X.sum(axis=0).A1
    adata.var['n_cells']=(adata.X>0).sum(axis=0).A1

    ## Print basic information about dataset
    if print_filtering_results==True:  
        logging.info(f"Dataset: {panel_sample_name} - {bays_model_name}")
        logging.info(f'Number of raw transcripts: {len(seg)}')
        logging.info(f"Number of transcripts deemed as noise: {sum(seg['cell'].isna())} - {round(pct_noise*100,2)}%")
        logging.info(f'Number of transcripts kept after filtering: {len(seg_true)} - {round(100*len(seg_true)/len(seg),2)}%')
        print('=================================')

    if plot_confidence_distr==True:
        fig_suptitle=' - '.join([panel_sample_name,bays_model_name])

        fig,ax=plt.subplots(1,2,figsize=(6,3.5))
        sns.histplot()
        sns.histplot(seg['confidence'],bins=100,ax=ax[0])
        sns.histplot(seg['assignment_confidence'],bins=100,ax=ax[1])
        ax[0].axvline(confidence_thr,color='red')
        ax[1].axvline(assignment_confidence_thr,color='red')

        fig.suptitle(fig_suptitle,fontweight='bold',y=1)
        plt.tight_layout()


    return adata   


### =================================================================
## Create function that splits a range of numbers into equal sized bins + adjusts the last bin's size, as it can be different from bin_size
#  This function is used to create the coordinates of a 2D bins, binning the background signal of Baysor outputs in order to add the bins to an
#  andnata object 
def split_range_to_bins(start,end,bin_size):
    bins = np.arange(start, end + bin_size, bin_size)
    
    # If the last bin is larger than the specified bin size, adjust it
    if bins[-1] > end:
        bins[-1] = end
    
    return bins


### =================================================================
## Bin the background signal not assigned to cells into rectangles with side length of 'bin_rect_side_length' & create anndata
#  extract the gene counts for each rectangle-shaped bin into an anndata (each row corresponds to one bin, columns are genes)
def bin_background_signal_into_anndata(bg,bin_rect_side_length):
    import time
    import warnings
    warnings.filterwarnings("ignore")

    '''
    @bg: dataframe containing only transcripts that were not assigned to a cell and wee not considered as noise in the Baysor output
    @bin_rect_side_length: Background gets binned into squares with side length==bin_rect_side_length -> int
    '''
        
    # bin the data into equally spaced groups   
    x_bins=split_range_to_bins(bg['x'].min().astype(int),bg['x'].max().astype(int)+1,bin_rect_side_length+1)
    y_bins=split_range_to_bins(bg['y'].min().astype(int),bg['y'].max().astype(int)+1,bin_rect_side_length+1)
    bg.loc[:,'x_cut']=pd.cut(bg['x'], x_bins, right=False,include_lowest=False)
    bg.loc[:,'y_cut']=pd.cut(bg['y'], y_bins, right=False,include_lowest=False)

    ## Extract names and counts of genes for each square bin of the background
    a=bg[['x','y','x_cut','y_cut','gene']].groupby(['x_cut', 'y_cut']).apply(lambda x: x['gene'].value_counts()).reset_index().set_index(['x_cut', 'y_cut'])

    
    ## Extract gene names and counts for each bin and transform it into an expression matrix format (count_df)
    l=[]
    for gene,bg_bin in a.reset_index().set_index('gene').groupby('gene'):
        count_series=bg_bin.reset_index().set_index(['x_cut','y_cut']).drop(columns=['gene'])
        count_series=count_series.rename(columns={'count': gene})
        l.append(count_series)
    
    count_df=pd.concat(l,axis=1,join='outer').fillna(0)

    
    ## Create adata.obs dataframe
    obs=pd.DataFrame(index=np.arange(len(count_df.index)))
    
    # Create adata.var dataframe
    var=pd.DataFrame(index=count_df.columns)
    
    
    #### CREATE ANNDATA OBJECT
    # Extract the count matrix of true cells from count_df
    from scipy import sparse
    X=count_df.values
    adata=ad.AnnData(X, obs=obs, var=var, dtype='float64')
    adata.X=sparse.csr_matrix(adata.X)

    # Calculate some basic statistics   
    adata.obs['n_counts']=adata.X.sum(axis=1).A1
    adata.obs['n_genes']=(adata.X>0).sum(axis=1).A1
    adata.var['n_counts']=adata.X.sum(axis=0).A1
    adata.var['n_cells']=(adata.X>0).sum(axis=0).A1


    ## Extract x,y coordinates of bins and save them as an np.array for plotting the bins later
    bin_coords=count_df.index.drop_duplicates()
    
    x_coords=pd.IntervalIndex(bin_coords.get_level_values('x_cut'))
    y_coords=pd.IntervalIndex(bin_coords.get_level_values('y_cut'))
    
    x_coords_arr=np.transpose(np.array([x_coords.left.tolist(),x_coords.left.tolist(),(x_coords.right-1).tolist(),(x_coords.right-1).tolist()]))
    y_coords_arr=np.transpose(np.array([y_coords.left.tolist(),(y_coords.right-1).tolist(),y_coords.left.tolist(),(y_coords.right-1).tolist()]))
    polygon_vertex_coords=np.transpose(np.stack((x_coords_arr,y_coords_arr)),(1,2,0))
    adata.uns['polygon_vertex_coords']=polygon_vertex_coords


    ## Calculate and add centroids of bins to adata.obs + 
    adata.obs['x_centroid']=((x_coords.left+x_coords.right-1)/2).values
    adata.obs['y_centroid']=((y_coords.left+y_coords.right-1)/2).values
    adata.obsm["spatial"] = adata.obs[["x_centroid", "y_centroid"]].copy().to_numpy()

    return adata

### =================================================================
## Calculate Pearson correlation matrix between the expression vectors of igh quality cells and the expression vector of their summed neighbourhood
def calculate_corr_with_neighbouring_bg(adata,n_neighbors,max_radius,coord_unit='pixels',consider_low_quality_cells_as_background=False):
    '''
    adata: anndata containing the background bins and the foreground (cells)
    n_neighbors: number of neigbours (==k) to consider when building kNN neighbourhood => set to gigher number in order to catch a decent number of background bins
    max_radius: only consider kNN-background neighbours that are within the distance of maximal radius to the given cell, unit: PIXELS!!!
    consider_low_quality_cells_as_background: boolean, whether to consider low quality cells as background => 
                                              if True, they will also be treated as background and therefore \
                                              will be used to construct the kNN background neighbourhood
    '''

    from sklearn.neighbors import NearestNeighbors
    import logging
    logging.basicConfig(format='%(asctime)s - %(message)s', datefmt='%d-%b-%y %H:%M:%S')
    logging.getLogger().setLevel(logging.INFO)
    from scipy import sparse
    import scipy

    logging.info(f"Calculating correlation of cells with surrounding background bins. Parameters: n_neighbors:{n_neighbors}, max_radius:{max_radius}")
    
    ##======== 1.
    ## Extract coordinates of high_quality_cells 
    cell_coords=adata.obs.loc[adata.obs['label_for_knn_search']=='high_qual_cell',['x','y']].values
    
    if consider_low_quality_cells_as_background==True:
        bg_cats=['background','low_qual_cell']
    else:
        bg_cats=['background']   
    bg_coords=adata.obs.loc[adata.obs['label_for_knn_search'].isin(bg_cats),['x','y']].values 
    
    ##======== 2.
    ## Fit a NearestNeighbors model on array 'background'
    nbrs=NearestNeighbors(n_neighbors=n_neighbors, algorithm='auto').fit(bg_coords)
    
    ## Find the k-nearest neighbors for each cell
    distances,indices=nbrs.kneighbors(cell_coords)

    ##======== 3.
    ## Mask those k-neares neigbours that are outside of the max_radius distance
    nbrs_inside_radius_mask=distances<=max_radius

    # Find the row indices with only False values ==> there were no neigbours within max_radius range
    indices_with_no_neighbors=np.where(np.all(nbrs_inside_radius_mask == False, axis=1))[0]
    logging.info(f"Number of high quality cells with no neighbours within {max_radius} {coord_unit} distance: {len(indices_with_no_neighbors)}")

    
    ##======== 4.
    ## For each cell sum the expression of the k-nearest neighbor and save it into a np.array 
    ngbr_expressions=np.array([c.X[indices[n][nbrs_inside_radius_mask[n]]].sum(axis=0) for n in range(indices.shape[0])])
    ngbr_expressions=np.squeeze(ngbr_expressions,axis=1)

    ##### DEPRECATED: CALCULATING THE MEAN EXPRESSION OF KNN NAIGHBOURHOOD DOESN'T CHANGE THE PEARSON CORRELATION, AS IT IS ONLY THE SUM/NUM_OF_BINS =>
    #####             DIVIDING BY SCALAR HAS NO EFFECT ON PEARSON CORRELATION
    ## Extract the number of NN background bins for each cell and use it to calculate the mean expression of the NN-expression
    #num_of_bg_bins_inside_radius=nbrs_inside_radius_mask.sum(axis=1)
    #ngbr_expressions_mean=ngbr_expressions/num_of_bg_bins_inside_radius[:, np.newaxis]
    #ngbr_expressions=ngbr_expressions_mean.copy()
    
    ## Calculate Pearson correlation matrix between the expression vectors for each high quality cell and the expression vector of its summed neighbourhood
    #corr_mat=np.corrcoef(adata.X[adata.obs['label_for_knn_search']=='high_qual_cell'].A,ngbr_expressions,rowvar=True) 
    cosine_dist=scipy.spatial.distance.cdist(adata.X[adata.obs['label_for_knn_search']=='high_qual_cell'].A,ngbr_expressions,metric='cosine')
    
    #  - np.corrcoef() calculates the correlation between all possible pairs of a the 2 2D-matrices 
    #  => extract only the Pearson correlation between high quality cells & its summed neighbourhoods
    #cell_corr_with_neighboring_bg=corr_mat[:cell_coords.shape[0],cell_coords.shape[0]:].diagonal()
    cell_corr_with_neighboring_bg=cosine_dist.diagonal()
    


    ##======== 5.
    ## Add Pearson correlations + to adata.obs
    adata.obs['Pearson_corr_with_bg_radius_'+str(max_radius)]=np.nan
    adata.obs.loc[adata.obs['label_for_knn_search']=='high_qual_cell','Pearson_corr_with_bg_radius_'+str(max_radius)]=cell_corr_with_neighboring_bg
    
    #adata.uns[f'{n_neighbors}_NN-neighbours_max_radius_{max_radius}_dict']={'distances':s_distances,
    #                                                                        'indices':indices,
    #                                                                        'indices_with_no_neighbors':indices_with_no_neighbors}

    adata.uns[f'{n_neighbors}_NN-neighbours_max_radius_{max_radius}_distances']=sparse.csr_matrix(distances)
    adata.uns[f'{n_neighbors}_NN-neighbours_max_radius_{max_radius}_indices']=indices
    adata.uns[f'{n_neighbors}_NN-neighbours_max_radius_{max_radius}_indices_with_no_neighbors']=indices_with_no_neighbors
    adata.uns[f'{n_neighbors}_NN-neighbours_max_radius_{max_radius}_hq_cells']=adata.obs.loc[adata.obs['label_for_knn_search']=='high_qual_cell'].index.tolist()
    adata.uns[f'{n_neighbors}_NN-neighbours_max_radius_{max_radius}_ngbr_expressions']=sparse.csr_matrix(ngbr_expressions)


    return adata

### =================================================================
## Split cells into low quality/high quality cells based on average_assignment_confidence and n_count thresholds
#  and add these labels as a new column ('label_for_knn_search')      
def define_low_high_quality_cells(c,avg_asg_thr,n_counts_quant_thr):
    #print(adata.obs['n_counts'].quantile(quant_thr))
    n_counts_thr=c.obs.loc[c.obs['origin']=='cell','n_counts'].quantile(n_counts_quant_thr)

    c.obs['label_for_knn_search']='high_qual_cell'

    c.obs.loc[(c.obs['origin']=='cell')\
              &((c.obs['avg_assignment_confidence']<=avg_asg_thr)\
                |(c.obs['n_counts']<=n_counts_thr))\
                 ,'label_for_knn_search']='low_qual_cell'

    c.obs.loc[c.obs['origin']=='background','label_for_knn_search']='background' 

    return c
    
```

## Extract "marker" genes of background Leiden clusters

## Calculate Pearson correlation of cells to their kNN background neigbourhood

```python
import warnings
warnings.filterwarnings('ignore')

data_dir="../../xenium_data/"#processed_data/cell_segmentation"

## Create panel_sample_names
panels=['Panel1','Panel2']
pat_nums=['P1','P2','P3','P4']
conditions=['D','H']
panel_sample_names=['_'.join([panel,pat_num,condition]) for panel in panels for pat_num in pat_nums for condition in conditions]

## List of Baysor segmentation methods that were created with these presegmentation methods ==> these methods will be looped over in script below
presegm_methods=['10x','no_segmentation','CPc','CPn']
#presegm_methods=['no_segmentation']

## Dict to collect the created adatas that contain Pearson correlation of cells to their kNN background neighbourhood
background_corr_adata_dict={}

## Loop through patient samples
for panel_sample_name in panel_sample_names[3:4]:
    background_corr_adata_dict[panel_sample_name]={}
    logging.info('===================')
    logging.info(panel_sample_name)
    logging.info('===================')

    baysor_out_fold=os.path.join(data_dir,'processed_data/baysor_output',panel_sample_name)
    bays_model_dirlist=[filename for filename in os.listdir(baysor_out_fold) if os.path.isdir(os.path.join(baysor_out_fold,filename))]
    
    bays_model_list=[bays_model for bays_model in bays_model_dirlist for presegm_meth in presegm_methods if presegm_meth in bays_model]
    bays_model_list.sort()
    
    
    for bays_model_name in bays_model_list[:]:
        logging.info('')
        logging.info(f'Model name:{bays_model_name}\n')
        
        ##===== READ NECESSARY INPUT FILES =======##
        ## Read in segmentation results + cell statistics output by Baysor (cells stats, area of cells)
        seg_fn=os.path.join(baysor_out_fold,bays_model_name,'segmentation.csv')
        seg=pd.read_csv(seg_fn)
    
        seg_stats_fn=os.path.join(baysor_out_fold,bays_model_name,'segmentation_cell_stats.csv')
        seg_stats=pd.read_csv(seg_stats_fn)
        
        ## Load the processed data as a dictionary
        proc_dir=os.path.join(data_dir,'processed_data/true_cell_filtering/baysor',panel_sample_name,bays_model_name)
        fpath=os.path.join(proc_dir,panel_sample_name+'.pickle')
        sample_dict=pickle.load(open(fpath, "rb"))
    
        cell_bound=sample_dict['cell_bound']
        cell_polygon_pixel_metrics=sample_dict['cell_polygon_pixel_metrics']
    
    
        ##===== GENERATE ADATA FROM BACKGROUND =======##
        bg=seg[(seg['cell'].isna()) & (seg['confidence']>0) &(seg['qv']>20)]
        
        ## Set the side length of the rectangle-shaped bins(in PIXELS!) ==> 
        #  - smaller binsize mean more bins (slower runtime) BUT makes the calculation of a more fine-grained kNN-neighbourhood possible (see in later step)
        bin_rect_side_length=50
        logging.info('Generating background bins')
        adata_bg=bin_background_signal_into_anndata(bg,bin_rect_side_length)
    
        
        ##===== GENERATE ADATA FROM CELLS =======##
        ## Filter for high-confidence transcripts and generate anndata object for sample
        logging.info('Generating anndata from cells')
        adata=generate_anndata(panel_sample_name,bays_model_name,seg,seg_stats,cell_polygon_pixel_metrics,confidence_thr=0.5,
                               assignment_confidence_thr=0.5,qv_thr=20,print_filtering_results=False,plot_confidence_distr=False)
    
    
        ##===== CONCATENATE BACKGROUND + FOREGROUND INTO ONE ADATA =======##
        import anndata as ad
        adata_bg.obs=adata_bg.obs.rename(columns={'x_centroid':'x','y_centroid':'y'})
        
        ## Concatenate adata of background bins + foreground(==cells)
        c=ad.concat([adata_bg,adata],join='outer',label='origin',uns_merge='unique',fill_value=0)
        
        ## Drop negative control transcripts + add origin as label to 'label_for_knn_search' to indicate background bins
        c=c[:,~c.var.index.str.contains('BLAN|Control',regex=True)]
        c.obs['origin']=c.obs['origin'].replace({'0':'background','1':'cell'},regex=True)
         

        ##===== SAVE ANNDATA =======##
        fpath=os.path.join(data_dir,'processed_data/true_cell_filtering/baysor/cell_corr_with_background',
                           panel_sample_name)
        
        if not os.path.exists(fpath):  
            os.makedirs(fpath)
            
        fname=bays_model_name.split('assignments_')[-1]+'_bg_corr_adata.h5ad'
        fn=os.path.join(fpath,fname)
        
        logging.info('Saving anndata file')
        c.write_h5ad(fn,compression='gzip')
        
        #background_corr_adata_dict[panel_sample_name][bays_model_name]=c


#### SAVE DICTIONARY WITH BACKGROUND ANNDATAS
#fpath=os.path.join(data_dir,'processed_data/true_cell_filtering/baysor','background_corr_adata_dict'+'.pickle')
#pickle.dump(background_corr_adata_dict, open(fpath, "wb"))
    
```

## Plot correlation of cells to their kNN background neigbourhood

```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.colors as mcolors
from scipy import sparse

data_dir="../../xenium_data/"#processed_data/cell_segmentation"

## Create panel_sample_names
panels=['Panel1','Panel2']
pat_nums=['P1','P2','P3','P4']
conditions=['D','H']
panel_sample_names=['_'.join([panel,pat_num,condition]) for panel in panels for pat_num in pat_nums for condition in conditions]

## List of Baysor segmentation methods that were created with these presegmentation methods ==> these methods will be looped over in script below
presegm_methods=['10x','no_segmentation','CPc','CPn']
presegm_methods=['10x']

corr_radius_df_dict={}

## Loop through patient samples
for panel_sample_name in panel_sample_names[3:4]:
    corr_radius_df_dict[panel_sample_name]={}

    baysor_out_fold=os.path.join(data_dir,'processed_data/baysor_output',panel_sample_name)
    bays_model_dirlist=[filename for filename in os.listdir(baysor_out_fold) if os.path.isdir(os.path.join(baysor_out_fold,filename))]
    
    bays_model_list=[bays_model for bays_model in bays_model_dirlist for presegm_meth in presegm_methods if presegm_meth in bays_model]
    bays_model_list.sort()
    
    for bays_model_name in bays_model_list[:]:
        logging.info(f"{' - '.join([panel_sample_name,bays_model_name])}\n")

        ##===== LOAD ANNDATA =======##
        fpath=os.path.join(data_dir,'processed_data/true_cell_filtering/baysor/cell_corr_with_background',
                           panel_sample_name)        
        fname=bays_model_name.split('assignments_')[-1]+'_bg_corr_adata.h5ad'
        fn=os.path.join(fpath,fname)
        c=sc.read_h5ad(fn)

        ## ===== DEFINE LOW AND HIGH QUALITY CELLS BASED ON THRESHOLDING ========
        ## Split cells into low quality/high quality cells based on average_assignment_confidence and n_count thresholds
        #  and add these labels as a new column ('label_for_knn_search')
        avg_asg_thr=0.5
        n_counts_quant_thr=0.1
        c=define_low_high_quality_cells(c,avg_asg_thr,n_counts_quant_thr) 


        ##===== CALCULATE CORRELATION BETWEEN CELL EXPRESSION AND THE SUMMED EXPRESSION VECTOR OF THEIR kNN BACKGROUND BIN-NEIGHBOURHOOD =======##
        ## Set the number of neighbors you want to find => set this parameter higher in order to find enough background bins
        #  => the max_radius parameter controls the number of kNN neighbours that will be considered
        n_neighbors=20
        
        ## Units: pixels!!
        max_radius_list=[300]
        for max_radius in max_radius_list[:]:
            c=calculate_corr_with_neighbouring_bg(c,n_neighbors,max_radius,
                                                  consider_low_quality_cells_as_background=False)
            logging.info('')

        
        adata_hq=c[c.obs['label_for_knn_search']=='high_qual_cell',:].copy()
        adata_hq.obs.loc[adata_hq.obs['cell_mask_area']==0,'cell_mask_area']=adata_hq.obs.loc[adata_hq.obs['cell_mask_area']==0,'area'].values
        adata_hq.layers['norm_by_area']=sparse.csr_matrix(adata_hq.layers['raw_counts'].multiply(1/(adata_hq.obs['cell_mask_area'].values.reshape(-1,1))))
        adata_hq.obsm["norm_by_area_pca"] = sc.tl.pca(adata_hq.layers["norm_by_area"],n_comps=20)
        sc.pp.neighbors(adata_hq,use_rep='norm_by_area_pca',n_pcs=20,metric='euclidean')
        #sc.tl.umap(adata_hq,random_state=0)
        sc.tl.leiden(adata_hq,resolution=0.3,key_added='leiden')

        
        l=[]
        for max_radius in max_radius_list:      
            l.append(adata_hq.obs.groupby('leiden').apply(lambda x: x['Pearson_corr_with_bg_radius_'+str(max_radius)]))
        corr_radius_df=pd.concat(l,axis=1)    
        corr_radius_df_dict[panel_sample_name][bays_model_name]=corr_radius_df

        
        ###====== PLOT CORRELATION WITH BACKGROUND ========###
        import math
        import matplotlib.colors as colors

        for max_radius in max_radius_list: 
            cols_list=['Pearson_corr_with_bg_radius_'+str(max_radius),'leiden']
            ncols=2
            nrows=math.ceil(ncols/len(cols_list))
            fig=plt.figure(figsize=(ncols*8,nrows*7))
            fig.suptitle(panel_sample_name+' - '+bays_model_name+' - max radius '+str(max_radius),fontweight='bold')
            
            for n,col in enumerate(cols_list):
                ax=fig.add_subplot(nrows,ncols,n+1)
                if pd.api.types.is_numeric_dtype(adata_hq.obs[col]):
                    normalize=colors.Normalize(vmin=adata_hq.obs[col].min(), vmax=adata_hq.obs[col].quantile(0.999))
                    p=ax.scatter(x=adata_hq.obs['x'],y=adata_hq.obs['y'],c=adata_hq.obs[col],s=0.2,alpha=1,norm=normalize)
                    plt.colorbar(p,ax=ax)
                else:
                    sns.scatterplot(data=adata_hq.obs,x='x',y='y',hue=col,s=2,ax=ax,alpha=1)
                
                
                circle_x_coord=int(ax.get_xlim()[1]*0.05)
                circle_y_coord=int(ax.get_ylim()[1]*0.05)
                cir=plt.Circle((circle_x_coord,circle_y_coord), max_radius, color='r',fill=True)
                ax.add_patch(cir)
                ax.text(circle_x_coord*0.5,circle_y_coord*0.5,f'Neighbourhood radius:{max_radius}')
                ax.invert_yaxis()
            
            
            
            #fig,ax=plt.subplots(1,1,figsize=(7,5))
            #ax.hist(adata_hq.obs['Pearson_corr_with_bg_radius_'+str(max_radius)],bins=200);
            #fig,axes=plt.subplots(1,1,figsize=(10,4.5))
            y_cols=['n_counts','avg_assignment_confidence']
            #for ax,y_col in zip(axes.flatten(),y_cols):
            sns.jointplot(data=adata_hq.obs,x='Pearson_corr_with_bg_radius_'+str(max_radius),y='n_counts',s=4,
                          #hue='avg_assignment_confidence',
                          xlim=(adata_hq.obs['Pearson_corr_with_bg_radius_'+str(max_radius)].quantile(0.0001),None),
                          ylim=(None,adata_hq.obs['n_counts'].quantile(0.999)))#,
                          #marginal_kws={'common_norm':False,'bw_adjust':3,'cut':0})#,ax=axes)
            
            fig,ax=plt.subplots(1,1,figsize=(12,5))
            fig.suptitle(panel_sample_name+' - '+bays_model_name+' - max radius'+str(max_radius),fontweight='bold')
            sns.violinplot(data=adata_hq.obs,y='Pearson_corr_with_bg_radius_'+str(max_radius),x='leiden',ax=ax)
            sns.stripplot(data=adata_hq.obs,y='Pearson_corr_with_bg_radius_'+str(max_radius),x='leiden',ax=ax,s=0.5,color='black',alpha=0.5)
        
        
        #fig,axes=plt.subplots(1,2,figsize=(16,8))
        #for ax,(orig,df) in zip(axes.flatten(),c.obs.groupby('origin')):        
        #    sns.scatterplot(data=df,x='x',y='y',hue='label_for_knn_search',s=2,ax=ax,alpha=1)
        #    ax.invert_yaxis()
    
        #fig,ax=plt.subplots(1,1,figsize=(12,12))
        #fig.suptitle(' - '.join([panel_sample_name,bays_model_name]))
        #sns.scatterplot(data=c.obs,x='x',y='y',hue='label_for_knn_search',s=2,ax=ax,alpha=1)
        #ax.invert_yaxis()

        
```
